# Cleaning Data

```python
sf_hb_df= sf_hb_df.dropna()
sf_hb_df.head()
```

```python
# Converting the data types to correct values
sf_hb_df['Date'] = pd.to_datetime(sf_hb_df['Date'], format='%m/%d/%Y')
sf_hb_df.head()
```

# Dropping unnecessary columns

## columns:
Date -> date

Descript -> description

Category -> category

Incident Code -> incident_code

PdDistrict -> pd_district

 

Create dayofweek column

```python
sf_cleaned_df = sf_hb_df[['Date', 'Descript', 'Category', 'Incident Code', 'PdDistrict']].copy()
sf_cleaned_df['dayofweek'] = sf_cleaned_df['Date'].dt.day_name()

sf_cleaned_df = sf_cleaned_df.rename(columns={'Date':'date', 'Descript':'description', 'Category':'category', 'Incident Code':'incident_code', 'PdDistrict':'pd_district'})

sf_cleaned_df.head()
```

```python
#checking the DF

value_counts = sf_cleaned_df['category'].value_counts()
print(value_counts)
```

# Sorting the Data

```python
# extract crime types 'Prostitution' into a its own DF
sf_hb_2df = sf_cleaned_df[sf_hb_df['Category'].str.contains("PROSTITUTION", na=False)]
sf_hb_2df.describe()
```

```python
# sorting Prostitution DF by dates in 2008
SanFran_prostitution_2008 = sf_hb_2df.loc[(sf_hb_2df['date'] >= '2008-12-01') & (sf_hb_2df['date'] <= '2008-12-31')].set_index('date')
SanFran_prostitution_2008.head()
```

```python
# sorting Prostitution DF by dates in 2017
SanFran_prostitution_2017 = sf_hb_2df.loc[(sf_hb_2df['date'] >= '2017-12-01') & (sf_hb_2df['date'] <= '2017-12-31')].set_index('date')
SanFran_prostitution_2017.head()
```

```python
# extract crime types 'Drug/narcotic' into a its own DF
sf_hb_3df = sf_cleaned_df[sf_hb_df['Category'].str.contains("DRUG/NARCOTIC", na=False)]
sf_hb_3df.describe()
```

```python
# sorting drugs DF by dates in 2008
SanFran_drugs_2008 = sf_hb_3df.loc[(sf_hb_3df['date'] >= '2008-12-01') & (sf_hb_3df['date'] <= '2008-12-31')].set_index('date')
SanFran_drugs_2008.head()
```

```python
# sorting drugs DF by dates in 2017
SanFran_drugs_2017 = sf_hb_3df.loc[(sf_hb_3df['date'] >= '2017-12-01') & (sf_hb_3df['date'] <= '2017-12-31')].set_index('date')
SanFran_drugs_2017.head()
```

# Saving data to csv

```python
#Reset indexes in dataframes
SanFran_drugs_2008 = SanFran_drugs_2008.reset_index()
SanFran_drugs_2017 = SanFran_drugs_2017.reset_index()

SanFran_prostitution_2008 = SanFran_prostitution_2008.reset_index()
SanFran_prostitution_2017 = SanFran_prostitution_2017.reset_index()
```

```python
#save to CSV file
SanFran_drugs_2008.to_csv("./Cleaned_Resources_NY_SanFran/SanFran_drugs_2008.csv", index=False)
SanFran_drugs_2017.to_csv("./Cleaned_Resources_NY_SanFran/SanFran_drugs_2017.csv", index=False)

SanFran_prostitution_2008.to_csv("./Cleaned_Resources_NY_SanFran/SanFran_prostitution_2008.csv", index=False)
SanFran_prostitution_2017.to_csv("./Cleaned_Resources_NY_SanFran/SanFran_prostitution_2017.csv", index=False)
```
