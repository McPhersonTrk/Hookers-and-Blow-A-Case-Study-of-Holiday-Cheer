# Imports

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
# import Path as path
```

# loading CSVs

```python
#NY drugs
ny_drugs_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_drugs_dec_2008.csv')
ny_drugs_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_drugs_dec_2017.csv')

#NY prostitution
ny_prostitution_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_prostitution_dec_2008.csv')
ny_prostitution_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_prostitution_dec_2017.csv')

#SF drugs
sf_drugs_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_drugs_2008.csv')
sf_drugs_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_drugs_2017.csv')

#SF prostitution
sf_prostitution_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_prostitution_2008.csv')
sf_prostitution_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_prostitution_2017.csv')
```

```python
display(ny_drugs_dec_2008.head())
display(ny_drugs_dec_2017.head())

display(ny_prostitution_dec_2008.head())
display(ny_prostitution_dec_2017.head())

display(sf_drugs_dec_2008.head())
display(sf_drugs_dec_2017.head())

display(sf_prostitution_dec_2008.head())
display(sf_prostitution_dec_2017.head())
```

# Graphs

## Graph 1: Total number of drug arrests per day

## Graph 2: Normalized number of drug arrests per day

## Graph 3: Total number of prostitution arrests per day

## Graph 4: Normalized number of prostitution arrests per day

```python
# Define the order of days
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Count occurrences for each day in both years
sortedbyday_ny_2008 = ny_drugs_dec_2008['dayofweek'].value_counts().reindex(day_order, fill_value=0)
sortedbyday_ny_2017 = ny_drugs_dec_2017['dayofweek'].value_counts().reindex(day_order, fill_value=0)
sortedbyday_sf_2008 = sf_drugs_dec_2008['dayofweek'].value_counts().reindex(day_order, fill_value=0)
sortedbyday_sf_2017 = sf_drugs_dec_2017['dayofweek'].value_counts().reindex(day_order, fill_value=0)

#formatting the bars
ind = np.arange(7)
width = 0.15

# Plotting
plt.figure(figsize=(10,8))

plt.bar(ind-width, sortedbyday_ny_2008, width, color="blue", label="New York 2008")
plt.bar(ind, sortedbyday_ny_2017, width, color="red", label="New York 2017")
plt.bar(ind+width, sortedbyday_sf_2008, width, color="orange", label="San Fransisco 2008")
plt.bar(ind+width*2, sortedbyday_sf_2017, width, color="green", label="San Fransisco 2017")


plt.legend(loc="best")

plt.title("Number of Drug Arrests per Day")
plt.xlabel("Day of Week")
plt.ylabel("Number of Arrests")
plt.xticks(ind + width / 4, day_order)
# Print the chart to the screen
plt.show()
```

```python
#Normalizing the data
total_arrests_ny_2008 = ny_drugs_dec_2008['dayofweek'].count()
total_arrests_ny_2017 = ny_drugs_dec_2017['dayofweek'].count()
total_arrests_sf_2008 = sf_drugs_dec_2008['dayofweek'].count()
total_arrests_sf_2017 = sf_drugs_dec_2017['dayofweek'].count()

#normalized vals
norm_ny_2008 = (sortedbyday_ny_2008/total_arrests_ny_2008)*100
norm_ny_2017 = (sortedbyday_ny_2017/total_arrests_ny_2017)*100
norm_sf_2008 = (sortedbyday_sf_2008/total_arrests_sf_2008)*100
norm_sf_2017 = (sortedbyday_sf_2017/total_arrests_sf_2017)*100

#formatting the bars
ind = np.arange(7)
width = 0.15

# Plotting
plt.figure(figsize=(10,8))

plt.bar(ind-width, norm_ny_2008, width, color="blue", label="New York 2008")
plt.bar(ind, norm_ny_2017, width, color="red", label="New York 2017")
plt.bar(ind+width, norm_sf_2008, width, color="orange", label="San Fransisco 2008")
plt.bar(ind+width*2, norm_sf_2017, width, color="green", label="San Fransisco 2017")


plt.legend(loc="best")

plt.title("Percentage of Drug Arrests per Day")
plt.xlabel("Day of Week")
plt.ylabel("Arrest %")
plt.xticks(ind + width / 4, day_order)
# Print the chart to the screen
plt.show()
```
