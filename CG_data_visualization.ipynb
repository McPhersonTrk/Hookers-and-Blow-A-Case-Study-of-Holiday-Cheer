# loading CSVs

```python
#NY drugs
ny_drugs_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_drugs_dec_2008.csv')
ny_drugs_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_drugs_dec_2017.csv')

#NY prostitution
ny_prostitution_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_prostitution_dec_2008.csv')
ny_prostitution_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_prostitution_dec_2017.csv')

#SF drugs
sf_drugs_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_drugs_2008.csv')
sf_drugs_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_drugs_2017.csv')

#SF prostitution
sf_prostitution_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_prostitution_2008.csv')
sf_prostitution_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_prostitution_2017.csv')
```

# Graphs


```python
# Define the order of days
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Count occurrences for each day in both years
sortedbyday_ny_2008 = ny_drugs_dec_2008['dayofweek'].value_counts().reindex(day_order, fill_value=0)
sortedbyday_ny_2017 = ny_drugs_dec_2017['dayofweek'].value_counts().reindex(day_order, fill_value=0)
sortedbyday_sf_2008 = sf_drugs_dec_2008['dayofweek'].value_counts().reindex(day_order, fill_value=0)
sortedbyday_sf_2017 = sf_drugs_dec_2017['dayofweek'].value_counts().reindex(day_order, fill_value=0)

# Plotting
plt.figure(figsize=(10,6))

plt.bar(day_order, sortedbyday_ny_2008, color="blue", label="New York 2008")
plt.bar(day_order, sortedbyday_ny_2017, color="red", label="New York 2017")
plt.bar(day_order, sortedbyday_sf_2008, color="orange", label="San Fransisco 2008")
plt.bar(day_order, sortedbyday_sf_2017, color="green", label="San Fransisco 2017")


plt.legend(loc="best")

plt.title("Number of Drug Arrests per Day")
plt.xlabel("Day of Week")
plt.ylabel("Number of Arrests")

# Print the chart to the screen
plt.show()
```
