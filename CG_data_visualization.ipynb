```python
# Plot our line that will be used to track the second country's traveling companion percentage over the years
plt.bar(ny_drugs_dec_2008['date'], ny_drugs_dec_2008['pd_district'], color="blue", label="2008")

# Plot our line that will be used to track the third country's traveling companion percentage over the years
# plt.bar(ny_drugs_dec_2017['date'], ny_drugs_dec_2017['pd_district'], color="orange", label="2017")

# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc="best")

plt.title("Date vs. Location")
plt.xlabel("Date")
# plt.xticks()
plt.ylabel("% District")

# Print our chart to the screen
plt.show()
```

```python
dayofweek = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
sortedbyday_2008 = ny_drugs_dec_2008['dayofweek'].value_counts()
sortedbyday_2017 = ny_drugs_dec_2017['dayofweek'].value_counts()


plt.figure(figsize=(10,6))

plt.bar(dayofweek, sortedbyday_2008, color="blue", label="New York 2008")
plt.bar(dayofweek, sortedbyday_2017, color="red", label="New York 2017")

plt.legend(loc="best")

plt.title("Number per Day")
plt.xlabel("Day of Week")
# plt.xticks()
plt.ylabel("Number")

# Print our chart to the screen
plt.show()
```
