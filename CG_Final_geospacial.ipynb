```python
import folium
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
#NY drugs
ny_drugs_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_drugs_dec_2008.csv')
ny_drugs_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_drugs_dec_2017.csv')

#NY prostitution
ny_prostitution_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_prostitution_dec_2008.csv')
ny_prostitution_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/ny_prostitution_dec_2017.csv')

#SF drugs
sf_drugs_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_drugs_2008.csv')
sf_drugs_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_drugs_2017.csv')

#SF prostitution
sf_prostitution_dec_2008 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_prostitution_2008.csv')
sf_prostitution_dec_2017 = pd.read_csv('./Cleaned_Resources_NY_SanFran/SanFran_prostitution_2017.csv')
```

# Maps by district New York City

```python
ny_lat = 40.7128
ny_long = -74.0060

# Create a map centered at a specific location
map_drugs_dec_2008 = folium.Map(location=[ny_lat, ny_long], zoom_start=10)
map_drugs_dec_2017 = folium.Map(location=[ny_lat, ny_long], zoom_start=10)
map_prostitution_dec_2008 = folium.Map(location=[ny_lat, ny_long], zoom_start=10)
map_prostitution_dec_2017 = folium.Map(location=[ny_lat, ny_long], zoom_start=10)

# for latitude, longitude in ny_drugs_dec_2008.iterrows():
#     folium.Marker(location=[latitude, longitude], popup='Marker Popup', icon=folium.Icon(color="green")).add_to(m)
i = 0

# Loop through the rows and add markers
for index, row in ny_drugs_dec_2008.iterrows():
    i=i+1
    if(i%20!=0):
        continue
    
    if(row['pd_district'] == 'B'):
        folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="green")).add_to(map_drugs_dec_2008)
    if(row['pd_district'] == 'Q'):
        folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="orange")).add_to(map_drugs_dec_2008)
    if(row['pd_district'] == 'S'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="darkblue")).add_to(map_drugs_dec_2008)
    if(row['pd_district'] == 'K'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="blue")).add_to(map_drugs_dec_2008)
    if(row['pd_district'] == 'M'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="red")).add_to(map_drugs_dec_2008)

for index, row in ny_drugs_dec_2017.iterrows():
    i=i+1
    if(i%10!=0):
        continue
    
    if(row['pd_district'] == 'B'):
        folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="green")).add_to(map_drugs_dec_2017)
    if(row['pd_district'] == 'Q'):
        folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="orange")).add_to(map_drugs_dec_2017)
    if(row['pd_district'] == 'S'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="darkblue")).add_to(map_drugs_dec_2017)
    if(row['pd_district'] == 'K'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="blue")).add_to(map_drugs_dec_2017)
    if(row['pd_district'] == 'M'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="red")).add_to(map_drugs_dec_2017)

for index, row in ny_prostitution_dec_2008.iterrows():
    if(row['pd_district'] == 'B'):
        folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="green")).add_to(map_prostitution_dec_2008)
    if(row['pd_district'] == 'Q'):
        folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="orange")).add_to(map_prostitution_dec_2008)
    if(row['pd_district'] == 'S'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="darkblue")).add_to(map_prostitution_dec_2008)
    if(row['pd_district'] == 'K'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="blue")).add_to(map_prostitution_dec_2008)
    if(row['pd_district'] == 'M'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="red")).add_to(map_prostitution_dec_2008)

for index, row in ny_prostitution_dec_2017.iterrows():    
    if(row['pd_district'] == 'B'):
        folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="green")).add_to(map_prostitution_dec_2017)
    if(row['pd_district'] == 'Q'):
        folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="orange")).add_to(map_prostitution_dec_2017)
    if(row['pd_district'] == 'S'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="darkblue")).add_to(map_prostitution_dec_2017)
    if(row['pd_district'] == 'K'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="blue")).add_to(map_prostitution_dec_2017)
    if(row['pd_district'] == 'M'):
            folium.Marker(location=[row['Latitude'], row['Longitude']], popup=f"Point {index}", icon=folium.Icon(color="red")).add_to(map_prostitution_dec_2017)
```
